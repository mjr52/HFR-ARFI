# Plot Initial HFR_ARFI Output

```python
# load in the data
from IPython import get_ipython
get_ipython().magic('reset -sf') 

import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np

matdata = sio.loadmat('res_sim.mat')
mirror = np.flip(matdata['arfidata'], 1)
data = np.concatenate((mirror, matdata['arfidata']), 1)

def plotfig(data, matdata, t):
    plt.figure(figsize=(5,5))
    plt.imshow(data[:,:,t], \
                   extent = [-matdata['lat'].max(), matdata['lat'].max(), matdata['axial'].max(), matdata['axial'].min()], \
              aspect = 'auto')
    plt.colorbar()
    plt.show()
    
plotfig(data,matdata, 1)
```

```python
plotfig(data, matdata, 10)
```

```python
plotfig(data, matdata, 20)
```

```python
plotfig(data, matdata, 30)
```

```python
plotfig(data, matdata, 36)
```
